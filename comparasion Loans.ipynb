```python
hdfc_loan['Bank'] = "HDFC Bank"
icici_loan['Bank'] = "ICICI Bank"
sbi_loan['Bank'] = 'SBI Bank'
axis_loan['Bank'] = 'Axis Bank'
```

```python
combined_df = pd.concat([axis_loan, sbi_loan, icici_loan, hdfc_loan])
combined_df.reset_index(drop=True)
```

```python
#Distribution of number of loans in all the banks.
import matplotlib.pyplot as plt
bank_counts = combined_df['Bank'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(bank_counts.index, bank_counts)
plt.xlabel('Bank Name')
plt.ylabel('Count')
plt.title('Distribution of Number of loans in Banks ')
plt.show()
```

```python
print("Select the type of Loan : 1: Home Loan, 2: Personal Loan, 3: Car Loan,4: Education Loan,5: Loan Against Property, 6: Business Loan, 7: Gold Loan,8: Agriculture Loan, 9: Vehicle Loan, 10: Top-Up Loan,11: Consumer Durable Loan, 12: Health Care Finance,13: Agricultural Loans, 14: Loan Against Securities")
LoanNeeded = int(input())
```

```python
LoanType = {1:'Home Loan', 2:'Personal Loan', 3:'Car Loan', 4:'Education Loan',
       5:'Loan Against Property',6:'Business Loan',7:'Gold Loan',
       8:'Agriculture Loan',9:'Vehicle Loan',10:'Top-Up Loan',
       11:'Consumer Durable Loan',12:'Health Care Finance',
       13:'Agricultural Loans',14:'Loan Against Securities'}
```

```python
print(f"You have selected : {LoanType[LoanNeeded]}")
```

```python
Loan = combined_df[combined_df["Loan Type"] == LoanType[LoanNeeded]]
Loan=Loan.reset_index(drop=True)
```

```python
Loan
```

```python
def age_in_range(row, user_age):
    age_range = row['Age Criteria']
    age_range = age_range.replace('years', '').split('-')
    min_age = int(age_range[0].strip())
    max_age = int(age_range[1].strip())
    return min_age <= user_age <= max_age
```

```python
user_age = int(input("Enter the user's age: "))

Loan = Loan[Loan.apply(lambda row: age_in_range(row, user_age), axis=1)]
Loan
```

```python
eleg = Loan['Eligibility'].unique().tolist()
LoanEleg = [item.strip() for sublist in eleg for item in sublist.split(",")]
LoanEleg = set(LoanEleg)
LoanEleg = list(LoanEleg)
```

```python
LoanEleg
```

```python
Loanele = {i+1: LoanEleg[i] for i in range(len(LoanEleg))}
Loanele
```

```python
print(f"Select any 2 categories to check for elegibility criteria: {Loanele}")
ele1 = int(input())
ele2 = int(input())
```

```python
Loan1 = Loan[Loan['Eligibility'].str.contains(Loanele[ele2], case=False)]
Loan2 = Loan[Loan['Eligibility'].str.contains(Loanele[ele1], case=False)]
```

```python
Loan= pd.concat([Loan1, Loan2]).drop_duplicates().reset_index(drop=True)
Loan
```

```python
feat1 = Loan['Features'].unique().tolist()
features = [item.strip() for sublist in feat1 for item in sublist.split(",")]
features = set(features)
features = list(features)
LoanFeatures = {i+1: features[i] for i in range(len(features))}
LoanFeatures
```

```python
print(f"Select any 2 most important feature according to your requirements \n{LoanFeatures}")
f1 = int(input())
f2 = int(input())
```

```python
Feat1 = Loan[Loan['Features'].str.contains(LoanFeatures[f1], case=False)]
```

```python
Feat2 = Loan[Loan['Features'].str.contains(LoanFeatures[f2], case=False, regex=False)]
```

```python
Loan= pd.concat([Feat1, Feat2]).drop_duplicates().reset_index(drop=True)
Loan
```
