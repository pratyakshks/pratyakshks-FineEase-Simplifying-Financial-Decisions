```python
import pandas as pd
```

```python
hdfc_bank = pd.read_csv("HDFC Accounts.csv", encoding='ISO-8859-1')
icici_bank = pd.read_csv("ICICI Accounts.csv")
axis_bank = pd.read_csv("Axis Bank Accounts.csv")
sbi_bank = pd.read_csv("SBI Accounts.csv")
```

```python
hdfc_bank['Bank'] = "HDFC Bank"
icici_bank['Bank'] = "ICICI Bank"
sbi_bank['Bank'] = 'SBI Bank'
axis_bank['Bank'] = 'Axis Bank'
```

```python
combined_df = pd.concat([axis_bank, sbi_bank, icici_bank, hdfc_bank])
```

```python
combined_df.info()
```

```python
combined_df.isnull().sum()
```

```python
saving_accounts = combined_df[combined_df['Account Type'].str.contains('Saving', case=False, na=False)]
saving_accounts = saving_accounts.reset_index(drop=True)
saving_accounts
```

```python
cleaned_column = saving_accounts['Age Criteria'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
value_counts = cleaned_column.value_counts()
value_counts
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize=(10, 6))
sns.countplot(data=saving_accounts, y='Account Type', order=saving_accounts['Account Type'].value_counts().index)
plt.title('Frequency of Each Account Type')
plt.xlabel('Count')
plt.ylabel('Account Type')
plt.show()
```

```python
#Distribution of Age Criteria in the Banks
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Distribution of Age Criteria in the Banks')
plt.show()
```

```python
#Distribution of saving accounts in various banks
bank_counts = saving_accounts['Bank'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(bank_counts.index, bank_counts)
plt.xlabel('Bank Name')
plt.ylabel('Count')
plt.title('Distribution of Saving Accounts in Banks ')
plt.show()
```
